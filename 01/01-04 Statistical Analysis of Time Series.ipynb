```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
def symbol_to_path(symbol,base_dir="data"):
    #return path to cvs
    return os.path.join(base_dir,"{}.csv".format(str(symbol)))
```

```python
def get_data(symbols,dates):
    df = pd.DataFrame(index=dates)
    if 'SPY' not in symbols: #add SPY for referanece
        symbols.insert(0,'SPY')
    
    for symbol in symbols:
        tmp_df = pd.read_csv(symbol_to_path(symbol),
                            usecols=['Date','Adj Close'],
                            na_values=['nan'],
                            index_col="Date",parse_dates=True)
        tmp_df = tmp_df.rename(columns={'Adj Close': symbol})
        df = df.join(tmp_df)
        if symbol == 'SPY':
            df = df.dropna(subset=['SPY'])
    return df
```

```python
def plot_data(df, title="Stock Prices"):
    ax = df.plot(title=title)
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()
```

```python
def plot_selected(df,columns,start_index,end_index):
    #given the start and end dates, plot the relevant info
    plot_data(df.ix[start_index:end_index,columns], title="Selected Data")
```

```python
def normalize_data(df):
    return df / df.ix[0,:]
```

```python
def global_stats_run():
    dates = pd.date_range('2008-11-05', '2009-11-25')
    symbols = ['SPY', 'GOOG', 'IBM']
    df = get_data(symbols, dates)
    plot_data(df)
    
    #compute global statistics for each stock
    print(df.mean())
    print(df.median())
    #standard deviation
    #a measure of how much it's changed over time/volitility
    print(df.std())
```

```python
if __name__ == "__main__":
    global_stats_run()
```

```python
def get_rolling_mean(values,window):
    return pd.rolling_mean(values,window=window)
```

```python
def get_rolling_std(values,window):
    return pd.rolling_std(values,window=window)
```

```python
def get_bollinger_bands(rm,rstd):
    upper_band = 2*rstd+rm
    lower_band = -2*rstd+rm
    return upper_band, lower_band
```

```python
def compute_daily_returns(df):
    daily_returns = (df / df.shift(1)) - 1
    daily_returns.ix[0,:] = 0
    return daily_returns
```

```python
def rolling_stats_run():
    dates = pd.date_range('2008-11-05', '2009-11-25')
    symbols = ['SPY']
    df = get_data(symbols, dates)
    
    #plot SPY data, retain plt axis object
    ax = df['SPY'].plot(title="SPY rolling mean", label='SPY')
    
    #compute rolling mean using a 3-day window
    rm_SPY = pd.rolling_mean(df['SPY'], window=3)
    
    #add rolling mean to the same plot
    rm_SPY.plot(label='Rolling mean', ax=ax)
    
    # Compute Bollinger Bands
    # 1. Compute rolling mean
    rm_SPY = get_rolling_mean(df['SPY'], window=20)

    # 2. Compute rolling standard deviation
    rstd_SPY = get_rolling_std(df['SPY'], window=20)
        
    # 3. Compute upper and lower bands
    upper_band, lower_band = get_bollinger_bands(rm_SPY, rstd_SPY)
    
    # Plot raw SPY values, rolling mean and Bollinger Bands
    ax = df['SPY'].plot(title="Bollinger Bands", label='SPY')
    rm_SPY.plot(label='Rolling mean', ax=ax)
    upper_band.plot(label='upper band', ax=ax)
    lower_band.plot(label='lower band', ax=ax)
    
    #add axis labels and legen
    ax.set_ylabel('Date')
    ax.set_xlabel('Price')
    ax.legend(loc='upper left')
    plt.show()
```

```python
if __name__ == "__main__":
    rolling_stats_run()
```
