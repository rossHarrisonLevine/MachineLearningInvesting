```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline   
from util import get_data, plot_data, compute_daily_returns
```

```python
def test():
    #read data
    dates = pd.date_range('2009-01-01', '2012-12-31')
    symbols = ['SPY']
    df = get_data(symbols,dates)
    plot_data(df)
    
    #compute daily returns
    daily_returns = compute_daily_returns(df)
    plot_data(daily_returns, title="Daily Returns over Time", ylabel='Daily Returns')
    
    #plot a histogram
    #daily_returns.hist()
    
    #has 10 bins by default, lets increase this with
    daily_returns.hist(bins=20)
    #should include a #plt.show() for real coding
    
    #now we need the mean and std deviation
    mean = daily_returns['SPY'].mean()
    print("mean = ", mean)
    std = daily_returns['SPY'].std()
    print("std = ", std)
          
    #add line for mean
    #all these commands will apply to the most recent graph created
    plt.axvline(mean, color='w', linestyle='dashed', linewidth=2)
    #pllot the positive std
    plt.axvline(std, color='r', linestyle='dashed', linewidth=2)
    #and the negative
    plt.axvline(-std, color='r', linestyle='dashed', linewidth=2)
    plt.show()
    
    #now for Kurtosis
    print("kurtosis = ",daily_returns["SPY"].kurtosis())
```

```python
if __name__ == '__main__':
    test()
```

```python
#plotting two histograms together
def multiple():
    #read data
    dates = pd.date_range('2009-01-01', '2012-12-31')
    symbols = ['SPY', 'XOM']
    df = get_data(symbols,dates)
    plot_data(df)
    
    #compute daily returns
    daily_returns = compute_daily_returns(df)
    plot_data(daily_returns, title="Daily Returns over Time", ylabel='Daily Returns')
    
    #plot histogram directly from dataframe on the same plot
    daily_returns['SPY'].hist(bins=20, label="SPY")
    daily_returns['XOM'].hist(bins=20, label="XOM")
```

```python
if __name__ == "__main__":
    multiple()
```
